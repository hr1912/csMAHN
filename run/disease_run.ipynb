````python
q_disease = 'LC,RA,Mtb,LN'.split(',')

df_para = pd.concat([pd.read_csv(p_cache.joinpath(
    'parameter_{}.csv'.format(i)
)) for i in q_disease])
df_para['path_ref'] = df_para['path_ref'].apply(lambda x:p_cache.joinpath(x))
df_para['path_que'] = df_para['path_que'].apply(lambda x:p_cache.joinpath(x))
display(df_para)

for _i, _row in df_para.iterrows():
    pass
_row
````

# RA

````python
disease = 'RA'
df_para_disease = df_para.query("tissue == '{}'".format(disease))

df_para_disease_info = df_para_disease.loc[:,'path_ref,path_que,name_ref,name_que'.split(',')]
df_para_disease_info['tag'] = df_apply_merge_field(df_para_disease_info,'{name_ref}-map-{name_que}')
df_para_disease_info = df_para_disease_info.drop(columns='name_ref,name_que'.split(','))

df_para_disease_info['obs_ref'] = df_para_disease_info['path_ref'].apply(
    lambda x:pd.read_csv(x.joinpath('obs.csv'),index_col=0))
df_para_disease_info['cell_count_ref'] = df_para_disease_info['obs_ref'].apply(
    lambda x:x.shape[0]
)
df_para_disease_info['_batch_ref'] = df_para_disease_info['obs_ref'].apply(
    lambda x:x['_batch'].unique().size
)
df_para_disease_info = df_para_disease_info.drop(columns='obs_ref'.split(','))
# que
df_para_disease_info['obs_que'] = df_para_disease_info['path_que'].apply(
    lambda x:pd.read_csv(x.joinpath('obs.csv'),index_col=0))
df_para_disease_info['cell_count_que'] = df_para_disease_info['obs_que'].apply(
    lambda x:x.shape[0]
)
df_para_disease_info['_batch_que'] = df_para_disease_info['obs_que'].apply(
    lambda x:x['_batch'].unique().size
)

df_para_disease_info = df_para_disease_info.drop(columns='obs_que'.split(','))
df_para_disease_info = df_para_disease_info.drop(columns='path_ref,path_que'.split(','))
display(df_para_disease_info)
````

````python
print('\n[run start] {}\n'.format(disease))

aligned=False
is_1v1 = False

n_epochs = [100, 100, 100]    
for _i, _row in df_para_disease.iterrows():
    run_cross_species_models(
        path_adata1=_row['path_ref'],
        path_adata2=_row['path_que'],
        key_class1=_row['key_cell_type'],
        key_class2=_row['key_cell_type'],
        sp1=_row['sp_simple_ref'],
        sp2=_row['sp_simple_que'],
        tissue_name=_row['tissue'],
        aligned=aligned,
        resdir_tag='{}-map-{};epochs={};is_1v1={}'.format(
            _row['name_ref'],
            _row['name_que'],
            sum(n_epochs),
            is_1v1),
        resdir=p_res,
        models='csMAHN,came'.split(','),
        **{'n_epochs': n_epochs,
           'is_1v1': is_1v1})

# SAMap
for _i, _row in df_para_disease.iterrows():
    run_cross_species_models(
        path_adata1=_row['path_ref'],
        path_adata2=_row['path_que'],
        key_class1=_row['key_cell_type'],
        key_class2=_row['key_cell_type'],
        sp1=_row['sp_simple_ref'],
        sp2=_row['sp_simple_que'],
        tissue_name=_row['tissue'],
        aligned=aligned,
        resdir_tag='{}-map-{};is_1v1={}'.format(
            _row['name_ref'],
            _row['name_que'],
            is_1v1),
        resdir=p_res,
        models='SAMap'.split(','),
        **{
           'is_1v1': is_1v1})


print('\n[run finish] {}\n'.format(disease))
````
